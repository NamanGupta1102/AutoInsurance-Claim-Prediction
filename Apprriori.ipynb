```python
def numerical2Range(dataframe, column_name: str):
    return pd.cut(dataframe[column_name], 20)
```

```python
def midValue(categories):
    return categories.apply(lambda x: x.mid)
```

```python
def replaceColumn(dataframe, column_name):
    dataframe[column_name] = midValue(numerical2Range(dataframe, column_name))
```

```python
replaceColumn(df, 'policy_tenure')
replaceColumn(df, 'age_of_car')
replaceColumn(df, 'age_of_policyholder')
```

```python
names_list = df.columns
print(names_list)
names =[]
for i in names_list: names.append(i)
print(names)
```

```python
def check(li,indices,data=data):
    count =0
    '''
    li = list of items to be checking count for
    indices = list of the indices postion 
    data = list of list of entries'''

    for d in data:
      flag = False
      for i in range(len(li)):
        if d[indices[i]] != li[i]:
          flag = True
          break
      if flag == False:
        count+=1
    return count
```

```python
def items(index,data=data):
  li = []
  for d in data:
    if d[index] not in li: li.append(d[index])
  return li
```

```python
frequency = []
for i in range(len(names)):
  key_values = dict()
  for x in items(i):
    key_values[x] = check([x],[i])
  frequency.append(key_values)
print(frequency)
```

```python
for i in range(len(frequency)):
  frequency[i] = sorted(frequency[i].items(), key=lambda x:x[1], reverse =True)

print(frequency)
```

```python
def create(frequency = frequency):
  ''' frequency is a list of list of most common used stuff'''
  combos = []

  def recc(li,depth):
    if depth == len(frequency): 
      combos.append(li.copy())
      print(li) # comment this later
      return
    for i in frequency[depth]:
      new_li = li.copy()
      new_li.append(i)
      recc(new_li.copy(),depth+1)

  li =[]
  recc(li.copy(),0)
  return combos
```

```python
final_combos = create()
```
